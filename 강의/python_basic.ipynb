    student = {"name": "John", "age": 20, "grade": "A"}

    student['weight'] = 50

    - 데이터 프레임 다룰 때 많이 사용

- 집합형(set)
    - 중복을 허용하지 않는 값들의 모음, 중괄호({})로 표현
    - 합집합, 교집합, 차집합 등의 연산 지원
    ```python
    unique_numbers = {1,2,3,4,5,5,5,5,5,5} # {1, 2, 3, 4, 5}
    ```
    - 중복값이 포함된 리스트의 유니크한 값만 뽑고 싶을 경우 사용
    - 합집합, 교집합, 차집합 연산을 할 때 사용

## 연산자

### 산술연산자
   - 덧셈 (`+`): 두 숫자를 더함
     - 예제:
       ```python
       result = 5 + 3  # 8
       ```
   - 뺄셈 (`-`): 두 숫자를 뺌
     - 예제:
       ```python
       result = 10 - 4  # 6
       ```
   - 곱셈 (`*`): 두 숫자를 곱함
     - 예제:
       ```python
       result = 7 * 2  # 14
       ```
   - 나눗셈 (`/`): 두 숫자를 나눔, 결과는 실수형
     - 예제:
       ```python
       result = 10 / 3  # 3.3333...
       ```
     - 0으로 나눌 수 없음 (에러 확인하는 법)
   - 몫 연산 (`//`): 두 숫자를 나누고 몫을 반환함 (정수형)
     - 예제:
       ```python
       result = 10 // 3  # 3
       ```
   - 나머지 연산 (`%`): 두 숫자를 나누고 나머지를 반환함
     - 예제:
       ```python
       result = 10 % 3  # 1
       ```
   - 거듭제곱 (`**`): 첫 번째 수를 두 번째 수만큼 제곱
     - 예제:
       ```python
       result = 2 ** 3  # 8
       ```

### 비교연산자
   - 두 값을 비교하고, 참(True) 또는 거짓(False) 값을 반환
   - 크다 (`>`): 왼쪽 값이 더 클 때 True 반환
     - 예제:
       ```python
       result = 5 > 3  # True
       ```
   - 작다 (`<`): 오른쪽 값이 더 클 때 True 반환
     - 예제:
       ```python
       result = 3 < 7  # True
       ```
   - 크거나 같다 (`>=`): 왼쪽 값이 크거나 같을 때 True 반환
     - 예제:
       ```python
       result = 5 >= 5  # True
       ```
   - 작거나 같다 (`<=`): 오른쪽 값이 크거나 같을 때 True 반환
     - 예제:
       ```python
       result = 4 <= 6  # True
       ```
   - 같다 (`==`): 두 값이 같을 때 True 반환
     - 예제:
       ```python
       result = 3 == 3  # True
       ```
   - 다르다 (`!=`): 두 값이 다를 때 True 반환
     - 예제:
       ```python
       result = 3 != 5  # True
       ```


### 논리연산자
   - and: 두 조건이 모두 참일 때만 True 반환
     - 예제:
       ```python
       result = (5 > 3) and (8 > 6)  # True
       ```
   - or: 두 조건 중 하나라도 참일 때 True 반환
     - 예제:
       ```python
       result = (5 > 3) or (2 > 6)  # True
       ```
   - not: 조건의 참/거짓을 반대로 바꿈
     - 예제:
       ```python
       result = not (5 > 3)  # False
       ```

### 할당연산자
   - 변수에 값을 할당하거나, 기존 값을 바탕으로 계산하여 다시 할당
   - 기본 할당 (`=`): 오른쪽 값을 왼쪽 변수에 할당
     - 예제:
       ```python
       x = 10
       ```
   - 더하고 할당 (`+=`): 변수에 값을 더한 후 다시 할당
     - 예제:
       ```python
       x += 5  # x = x + 5 (결과: 15)
       ```
   - 빼고 할당 (`-=`): 변수에서 값을 뺀 후 다시 할당
     - 예제:
       ```python
       x -= 3  # x = x - 3 (결과: 12)
       ```
   - 곱하고 할당 (`*=`): 변수에 값을 곱한 후 다시 할당
     - 예제:
       ```python
       x *= 2  # x = x * 2 (결과: 24)
       ```
   - 나누고 할당 (`/=`): 변수에 값을 나눈 후 다시 할당
     - 예제:
       ```python
       x /= 4  # x = x / 4 (결과: 6.0)
       ```
   - 나머지 할당 (`%=`): 변수에 나머지 값을 할당
     - 예제:
       ```python
       x %= 5  # x = x % 5 (결과: 1.0)
       ```
   - 몫 할당 (`//=`): 변수에 몫을 할당
     - 예제:
       ```python
       x //= 2  # x = x // 2 (결과: 0.0)
       ```
   - 거듭제곱 할당 (`=`): 변수에 거듭제곱 값을 할당
     - 예제:
       ```python
       x = 3  # x = x  3
       ```

### 비트연산자
   - 숫자의 이진수 표현을 기준으로 수행하는 연산
   - AND (`&`): 두 비트가 모두 1일 때 1 반환
   - OR (`|`): 두 비트 중 하나라도 1일 때 1 반환
   - NOT (`~`): 모든 비트를 반전
   - 데이터 프레임 다룰 때 많이 사용

### 연산자우선순위
- 괄호 > 지수(**) > 곱셈/나눗셈 > 덧셈/뺄셈 > 비교연산 > 논리연산

```python
result = (5 + 2) * 3  # 21 괄호로 우선순위 명시
```

## 제어문

### 조건문 (if,elif,else)
   - 조건문의 역할: 특정 조건에 따라 코드의 흐름을 제어
   - if 문 기본 구조: 조건이 참일 때만 해당 코드 블록을 실행
     - 예제:
       ```python
       age = 20
       if age >= 18:
           print("You are an adult.")
       ```
   - elif 문: 여러 조건을 검사할 때 사용, 첫 번째 조건이 False일 경우 다음 조건 검사
     - 예제:
       ```python
       score = 85
       if score >= 90:
           print("Grade A")
       elif score >= 80:
           print("Grade B")
       else:
           print("Grade C")
       ```
   - else 문: 앞의 모든 조건이 False일 때 실행되는 블록
     - 예제:
       ```python
       temperature = 10
       if temperature > 30:
           print("It's hot!")
       elif temperature > 20:
           print("It's warm!")
       else:
           print("It's cold!")
       ```
   - 중첩 조건문: 조건문 안에 또 다른 조건문을 작성 가능
     - 예제:
       ```python
       number = 5
       if number > 0:
           if number % 2 == 0:
               print("Even positive number")
           else:
               print("Odd positive number")
       else:
           print("Negative number")
       ```

### 반복문 (for)
   - for 문 기본 구조: 리스트나 문자열 같은 순차 자료형을 순회
     - 예제:
       ```python
       fruits = ["apple", "banana", "cherry"]
       for fruit in fruits:
           print(fruit)
       ```
   - range() 함수와 함께 사용: 특정 횟수만큼 반복 실행
     - 예제:
       ```python
       for i in range(5):
           print(i)  # 0부터 4까지 출력
       ```
   - 리스트 내포 (List Comprehension): 반복문을 사용해 리스트를 간결하게 생성
     - 예제:
       ```python
       squares = [x2 for x in range(5)]
       print(squares)  # [0, 1, 4, 9, 16]
       ```

### 반복문 (while)
   - while 문 기본 구조: 조건이 참일 동안 계속해서 반복
     - 예제:
       ```python
       count = 0
       while count < 5:
           print(count)
           count += 1
       ```
   - 무한 루프: 조건이 항상 참일 경우, 무한히 반복
     - 예제:
       ```python
       while True:
           print("This loop will run forever.")
           break  # 무한 루프를 방지하기 위해 break 사용
       ```

## 함수

   ---
   - 함수의 정의: 코드의 재사용성과 가독성을 높이기 위해 특정 작업을 묶어두는 코드 블록
   - 함수를 사용하는 이유: 중복 코드 제거, 코드의 유지보수성 증가, 가독성 향상
   - 함수 호출: 함수를 정의한 후, 함수 이름을 사용해 호출하여 실행
     - 예제:
       ```python
       def greet():
           print("Hello, Python!")

       greet()  # 함수 호출
       ```

 ---
   - 함수 정의: `def` 키워드를 사용하여 함수 정의
   - 기본 구조: 함수 이름과 매개변수를 사용해 함수 정의, `return`을 통해 값 반환 가능
     - 예제:
       ```python
       def add(a, b):
           return a + b

       result = add(5, 3)
       print(result)  # 8
       ```
   - 매개변수 (Parameters): 함수가 입력으로 받는 값
   - 인자 (Arguments): 함수 호출 시 함수에 전달되는 실제 값


---
   - 인자는 함수 정의에서 매개변수의 순서대로 전달
     - 예제:
       ```python
       def multiply(x, y):
           return x * y

       result = multiply(4, 5)  # 위치에 맞춰 값이 전달됨
       print(result)  # 20
       ```


---
   - 함수 정의 시 기본값을 설정하면, 호출할 때 값을 생략 가능
     - 예제:
       ```python
       def greet(name="Guest"):
           print(f"Hello, {name}!")

       greet()  # 기본값 'Guest' 사용
       greet("Alice")  # 'Alice' 전달
       ```



---
   - 매개변수 이름을 명시하여 인자를 전달할 수 있어, 순서와 상관없이 값을 할당
     - 예제:
       ```python
       def divide(numerator, denominator):
           return numerator / denominator

       result = divide(denominator=2, numerator=10)
       print(result)  # 5.0
       ```



---
   - 함수는 `return`을 사용해 값을 반환 가능, `return`이 없으면 기본적으로 `None`을 반환
   - 단일 반환값: 하나의 값을 반환
     - 예제:
       ```python
       def square(x):
           return x ** 2

       result = square(4)
       print(result)  # 16
       ```
   - 다중 반환값: 여러 값을 튜플 형태로 반환 가능
     - 예제:
       ```python
       def get_dimensions():
           width = 10
           height = 20
           return width, height

       w, h = get_dimensions()
       print(w, h)  # 10 20
       ```





---
   - 람다 함수란?: 짧고 간결한 함수 정의가 필요할 때 사용
   - 구조: `lambda` 키워드를 사용해 익명 함수 생성
     - 예제:
       ```python
       add = lambda a, b: a + b
       result = add(3, 5)
       print(result)  # 8
       ```
   - 람다 함수의 활용: `map()`, `filter()` 같은 함수와 함께 자주 사용
     - 예제:
       ```python
       numbers = [1, 2, 3, 4, 5]
       squared = list(map(lambda x: x  2, numbers))
       print(squared)  # [1, 4, 9, 16, 25]
       ```
