```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/under5MortalityRate.csv')
```

```python
df
```

```python
# df.drop('Indicator', axis=1, inplace=True)
df = df.drop(columns='Indicator')
df
```

```python
df['First Tooltip'] = df['First Tooltip'].map(lambda x : float(x.split("[")[0]))  # 실수 바꿔줌
# df['First Tooltip'] = df['First Tooltip'].str[0:5]
df
```

```python
cond1 = (df['Period'] >= 2015)
cond2 = (df['Dim1'] == 'Both sexes')

df1 = df.loc[cond1&cond2]
df1
```

```python
df1 = df1.pivot(index='Location', columns='Period', values='First Tooltip')
df1
```

```python
df2 = df.pivot_table(index='Dim1', columns='Period', values='First Tooltip')  
# pivot_table 은 pivot 과 달리 중복값 처리, 그룹바이, 집계처리 가능 
```
