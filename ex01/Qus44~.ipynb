```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv')
df.head()
```

```python
# df.groupby('host_name').count()
# df.groupby('host_name').size()
```

```python
df['host_name'].value_counts().sort_index() # 값 항목수 계산 후 인덱스로 정렬
```

```python
# df['host_name'].value_counts().to_frame().rename(columns= {'count' : 'counts'})
# df['host_name'].value_counts().to_frame('counts')
df.groupby('host_name').size().to_frame('counts').sort_values('counts', ascending=False)
```

```python
df.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size()
```

```python
# df.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size().groupby('neighbourhood_group', as_index=False).max()
df.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size().sort_values(['neighbourhood_group', 'size'], ascending=[True,False]).drop_duplicates('neighbourhood_group', keep='first')
```

```python
# df.groupby('neighbourhood_group')['price'].describe()
df.groupby('neighbourhood_group')['price'].agg(['mean', 'var', 'min', 'max'])
```

```python
# df.groupby('neighbourhood_group')['reviews_per_month'].agg(['mean', 'var', 'min', 'max'])
df[['neighbourhood_group','reviews_per_month']].groupby('neighbourhood_group').agg(['mean','var','max','min'])
```

```python
df.groupby(['neighbourhood', 'neighbourhood_group'])['price'].mean()
```

```python
# df.groupby(['neighbourhood', 'neighbourhood_group'])['price'].mean().unstack()
df.groupby(['neighbourhood', 'neighbourhood_group'], as_index=False)['price'].mean()
```

```python
df.groupby(['neighbourhood', 'neighbourhood_group'])['price'].mean().unstack().fillna(-999)
```

```python
df.loc[df['neighbourhood_group'] == 'Queens'].groupby('neighbourhood')['price'].agg(['mean', 'var', 'min', 'max'])
```
