```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/timeTest.csv')
df.info()
```

```python
df['Yr_Mo_Dy'] = pd.to_datetime(df['Yr_Mo_Dy'])
df
```

```python
df['Yr_Mo_Dy'].dt.year.unique()
```

```python
def fix_century(x):
    import datetime
    
    year = x.year - 100 if x.year >= 2061 else x.year
    return pd.to_datetime(datetime.date(year, x.month, x.day))  # datetime.date 형식을 pandas.datetime으로 변환 -> dataframe에 잘 들어가기 때문


df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(fix_century)

Ans = df.head(4)
```

```python
df.loc[df['Yr_Mo_Dy'].dt.year >= 2061, 'Yr_Mo_Dy'] = \
    df.loc[df['Yr_Mo_Dy'].dt.year >= 2061, 'Yr_Mo_Dy'] - pd.DateOffset(years=100)

df
```

```python
Ans = df.groupby(df['Yr_Mo_Dy'].dt.year).mean()
Ans
```

```python
df
```

```python
df['weekday'] = df['Yr_Mo_Dy'].dt.weekday       # 월요일 0 ~ 일요일 6
df['weekday'].head(3).to_frame()
```

```python
df['WeekCheck'] = df['weekday'].map(lambda x : 1 if (x == 5) | (x == 6) else 0)
df['WeekCheck'].head(3).to_frame()
```

```python
Ans = df.groupby(df['Yr_Mo_Dy'].dt.month, as_index=True).mean()
Ans
```

```python
df = df.fillna(method='ffill').fillna(method='bfill')
# 결측치에 직전값 대체 후 결측치에 직후값 대체
```
